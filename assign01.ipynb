### Sample code to get a (P,C) pair

```python
plain_text = "lasagna is better than pasta"
print("PT:",plain_text)

key1 = "00000000069f6bc7"   # 0th key
print("K1: ", key1)

k1 = des(binascii.unhexlify(key1), CBC, "\0\0\0\0\0\0\0\0", pad=None, padmode=PAD_PKCS5)

x = k1.encrypt(plain_text)
print ("1st E - X: %r" % x)


key2 = "00000000069f92d6"   # 9999th key
print("K2: ", key2)

k2 = des(binascii.unhexlify(key2), CBC, "\0\0\0\0\0\0\0\0", pad=None, padmode=PAD_PKCS5)
c = k2.encrypt(x)
print ("2nd E - C: %r" % c)


x = k2.decrypt(c)
plain_text = k1.decrypt(x)
print ("1st D - X: %r" % x)
print ("2nd D - PT: %r" % plain_text)
```

```python
def get_keys():
    keys= []
    for i in range(111111111,122222222):   # 11,111,111 keys
        i = hex(i)
        val = i[2:]
        
        for i in range(16):
            if len(val) < 16:
                val = '0' + val
        
        keys.append(val)
        
    return keys
```

```python
%time keys = get_keys()
```

```python
def MitM(keys, plain_text, cipher):
    count = 0
    
    for i in range(len(keys)):
        print("Step: ", i)
        key1 = keys[i]
        
        k1 = des(binascii.unhexlify(key1), CBC, "\0\0\0\0\0\0\0\0", pad=None, padmode=PAD_PKCS5)
        x_p = k1.encrypt(plain_text)
        
        for j in range(len(keys)):
            count += 1
            key2 = keys[j]
            print("Key 2:", key2)

            k2 = des(binascii.unhexlify(key2),CBC, "\0\0\0\0\0\0\0\0", pad=None, padmode=PAD_PKCS5)
            x_c = k2.decrypt(cipher)
            
            if x_p == x_c:
                return (key1, key2, count)
            else:
                continue
    
```

### Lookup tables

```python
def lookup_table(keys, plain_text, cipher):
    dic1, dic2 = {}, {}
    for i in range(len(keys)):
        key1 = keys[i]
        
        k1 = des(binascii.unhexlify(key1), CBC, "\0\0\0\0\0\0\0\0", pad=None, padmode=PAD_PKCS5)
        x_p = k1.encrypt(plain_text)
        dic1[key1] = x_p
        
        k2 = des(binascii.unhexlify(key1),CBC, "\0\0\0\0\0\0\0\0", pad=None, padmode=PAD_PKCS5)
        x_c = k2.decrypt(cipher)
        dic2[key1] = x_c
        
    return (dic1, dic2)
```

```python
%time PT_lookup, C_lookup = lookup_table(keys[0:100000], "lasagna is better than pasta",b'\x95]\xac77\x00\xb9tV\x83nC\xfb\x95\x95\xfb#\x99\xc3F\x8e;\xae\x9f\x97o\xf1\xbcq\xef\x06\x82F\x00\xedc\xeb\x97\x17U')
```
